```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import koreanize_matplotlib
from keras.models import Sequential
from keras.layers import LSTM, Dense

# 종목 데이터 가져오기
ticker = "053280.KS"
data = yf.download(ticker, start="2021-01-01", end="2021-11-19")['Adj Close']

# NaN 값 제거
data = data.dropna()

# 데이터 스케일링
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data.values.reshape(-1, 1))

# LSTM 데이터셋 생성 함수
def create_dataset(dataset, time_step=60):
    X, y = [], []
    for i in range(time_step, len(dataset)):
        X.append(dataset[i-time_step:i, 0])
        y.append(dataset[i, 0])
    return np.array(X), np.array(y)

# 예측할 기간 설정
forecast_start = "2021-10-11"
forecast_end = "2021-11-19"

forecast_index = pd.date_range(start=forecast_start, end=forecast_end, freq='B')  # 평일만 포함
forecast_days = len(forecast_index)

# 최적 train_ratio 찾기
best_mse = float('inf')
best_ratio = None
performance_results = []

for train_ratio in range(60, 95, 5):  # 60%부터 90%까지 5%씩 증가
    train_size = int(len(scaled_data) * (train_ratio / 100))
    test_size = len(scaled_data) - train_size

    # Train/Test 데이터 분리
    train_data = scaled_data[:train_size]
    test_data = scaled_data[train_size-60:]  # 과거 60일 포함

    # Train/Test 데이터셋 생성
    X_train, y_train = create_dataset(train_data)
    X_test, y_test = create_dataset(test_data)

    # 데이터 Reshape (LSTM 입력 형식)
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    # LSTM 모델 정의
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
        LSTM(50),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')

    # 모델 학습
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)

    # 예측
    y_pred = model.predict(X_test)
    y_pred = scaler.inverse_transform(y_pred)
    y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

    # 성능 지표 계산
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # 결과 저장
    performance_results.append((train_ratio, mse, mae, r2))
    print(f"Train {train_ratio}% -> MSE: {mse}, MAE: {mae}, R2: {r2}")

    # 최적 비율 갱신
    if mse < best_mse:
        best_mse = mse
        best_ratio = train_ratio

# 최적 Train 비율 출력
print(f"\n최적 Train 비율: {best_ratio}%")
print(f"최적 MSE: {best_mse}")

# 최적 비율로 예측 수행
train_size = int(len(scaled_data) * (best_ratio / 100))
train_data = scaled_data[:train_size]

X_train, y_train = create_dataset(train_data)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=0)

# 예측 준비 (마지막 학습 데이터에서 시작)
test_data = scaled_data[-60:]
test_data = np.array([test_data[:, 0]])
test_data = np.reshape(test_data, (test_data.shape[0], test_data.shape[1], 1))

# 10/11~11/19 예측 수행
predictions = []
for _ in range(forecast_days):
    next_pred = model.predict(test_data)[0][0]
    predictions.append(next_pred)
    test_data = np.append(test_data[:, 1:, :], [[[next_pred]]], axis=1)

# 예측 결과 복원
predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

# 실제 주가 가져오기 (10/11~11/19)
actual_data = yf.download(ticker, start=forecast_start, end=forecast_end)['Adj Close']

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(forecast_index, predictions, label="예측 값 (10/11~11/19)", color="orange")
plt.plot(actual_data.index, actual_data.values, label="실제 값 (10/11~11/19)", color="blue")
plt.title(f"{ticker} 주가 예측 vs 실제 (10/11~11/19)")
plt.xlabel("날짜")
plt.ylabel("주가")
plt.legend()
plt.show()

# 최종 예측 결과 DataFrame 출력
forecast_results = pd.DataFrame({'Date': forecast_index, 'Forecasted': predictions.flatten()})
forecast_results.set_index('Date', inplace=True)
forecast_results
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# 데이터 로드
file_path = '/Users/leesangwon/Downloads/코스피,코스닥 이상치 가설검정 위한 예측,실제 주가/코스이상치.csv'
tickers_data = pd.read_csv(file_path, encoding='utf-8')

# 결과를 저장할 데이터프레임
forecast_results_all = pd.DataFrame()

# LSTM 데이터셋 생성 함수
def create_dataset(dataset, time_step=60):
    X, y = [], []
    for i in range(time_step, len(dataset)):
        X.append(dataset[i-time_step:i, 0])
        y.append(dataset[i, 0])
    return np.array(X), np.array(y)

# 예측할 기간 설정
forecast_start = "2023-10-11"
forecast_end = "2023-11-19"

forecast_index = pd.date_range(start=forecast_start, end=forecast_end, freq='B')  # 평일만 포함
forecast_days = len(forecast_index)

# 종목별 예측 실행
for _, row in tickers_data.iterrows():
    ticker = row['Original_Ticker']
    print(f"Processing {ticker}...")

    # 주가 데이터 가져오기
    data = yf.download(ticker, start="2023-01-01", end="2023-11-19")['Adj Close']

    # 데이터 비어 있으면 스킵
    if data.empty:
        print(f"{ticker}: No data available, skipping.")
        continue

    # NaN 값 제거
    data = data.dropna()

    # 데이터 크기 확인
    if len(data) < 60:
        print(f"{ticker}: Not enough data for training, skipping.")
        continue

    # 데이터 스케일링
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data.values.reshape(-1, 1))

    # Train/Test 데이터 생성
    train_size = int(len(scaled_data) * 0.8)  # 80%를 학습 데이터로 사용
    train_data = scaled_data[:train_size]

    if len(train_data) < 60:  # 학습 데이터가 time_step보다 작으면 스킵
        print(f"{ticker}: Not enough train data for LSTM, skipping.")
        continue

    X_train, y_train = create_dataset(train_data)
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

    # LSTM 모델 정의
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
        LSTM(50),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')

    # 모델 학습
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)

    # 예측 준비 (마지막 학습 데이터에서 시작)
    test_data = scaled_data[-60:]
    if len(test_data) < 60:  # 예측 데이터가 time_step보다 작으면 스킵
        print(f"{ticker}: Not enough test data for prediction, skipping.")
        continue

    test_data = np.array([test_data[:, 0]])
    test_data = np.reshape(test_data, (test_data.shape[0], test_data.shape[1], 1))

    # 예측 수행
    predictions = []
    for _ in range(forecast_days):
        next_pred = model.predict(test_data)[0][0]
        predictions.append(next_pred)
        test_data = np.append(test_data[:, 1:, :], [[[next_pred]]], axis=1)

    # 예측 결과 복원
    predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

    # 결과 데이터프레임 생성
    forecast_df = pd.DataFrame({'Date': forecast_index, f'{ticker} Forecasted': predictions.flatten()})
    forecast_df.set_index('Date', inplace=True)

    # 결과 통합
    if forecast_results_all.empty:
        forecast_results_all = forecast_df
    else:
        forecast_results_all = pd.merge(forecast_results_all, forecast_df, left_index=True, right_index=True, how='outer')

# 예측 결과 CSV로 저장
forecast_results_all.to_csv('/Users/leesangwon/Downloads/forecast_results_all.csv')
print("All forecast results saved to 'forecast_results_all.csv'")
```

# ARIMA

```python
import numpy as np
import pandas as pd
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 종목 데이터 가져오기
ticker = "016450.KS"
data = yf.download(ticker, start="2024-01-01", end="2024-11-19")['Adj Close']

# NaN 값 제거
data = data.dropna()

# 예측할 기간 설정
forecast_start = "2024-10-11"
forecast_end = "2024-11-19"

# 실제 주가 가져오기
actual_data = yf.download(ticker, start=forecast_start, end=forecast_end)['Adj Close']
forecast_days = len(actual_data)

# ARIMA 모델 설정 및 학습
train_data = data[:-forecast_days]
test_data = data[-forecast_days:]

# 모델 학습
arima_order = (5, 1, 0)  # ARIMA(p, d, q)
model = ARIMA(train_data, order=arima_order)
fitted_model = model.fit()

# 예측
forecast = fitted_model.forecast(steps=forecast_days)
forecast_index = pd.date_range(start=forecast_start, end=forecast_end, freq='B')

# 성능 지표 계산
mse = mean_squared_error(test_data, forecast)
mae = mean_absolute_error(test_data, forecast)
r2 = r2_score(test_data, forecast)

print(f"ARIMA 모델 성능:\nMSE: {mse}, MAE: {mae}, R2: {r2}")

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(test_data.index, test_data.values, label="실제 값 (10/11~11/19)", color="blue")
plt.plot(forecast_index, forecast, label="예측 값 (10/11~11/19)", color="orange")
plt.title(f"{ticker} 주가 예측 vs 실제 (10/11~11/19) - ARIMA")
plt.xlabel("날짜")
plt.ylabel("주가")
plt.legend()
plt.show()

# 예측 결과 저장
forecast_results = pd.DataFrame({'Date': forecast_index, 'Forecasted': forecast})
forecast_results.set_index('Date', inplace=True)
forecast_results
```
