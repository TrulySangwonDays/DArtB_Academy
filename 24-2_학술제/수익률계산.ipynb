```python
import yfinance as yf

ticker = '005930.KS'

# 2024년 10월 9일부터 2024년 10월 11일까지의 데이터 가져오기
stock_data = yf.download(ticker, start='2024-10-09', end='2024-10-11', interval='1d')

# 종가(Adj Close) 데이터에서 일일 수익률 계산
stock_data['Daily Return'] = stock_data['Adj Close'].pct_change()

# 2024년 10월 11일의 수익률 출력
oct_11_return = stock_data.loc['2024-10-11', 'Daily Return']
print(f"2024년 10월 11일의 일일 수익률: {oct_11_return:.4f}")
```
