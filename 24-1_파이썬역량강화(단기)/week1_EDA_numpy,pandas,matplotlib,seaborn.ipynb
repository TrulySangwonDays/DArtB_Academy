```python
# 박스플롯이해하기
origin_desc = df.groupby('origin')['mpg'].describe()
```

```python
europe = origin_desc.loc['europe']
europe
```

```python
Q3 = europe['75%']
Q1 = europe['25%']
IQR = Q3 - Q1
OUT_MAX = Q3 + (1.5 * IQR)
OUT_MIN = Q1 - (1.5 * IQR)
OUT_MAX, OUT_MIN
```
